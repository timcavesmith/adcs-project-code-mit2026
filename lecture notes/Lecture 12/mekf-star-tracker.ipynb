```julia
using LinearAlgebra
using PythonPlot
```

```julia
#quaternion functions
function hat(v)
    return [0 -v[3] v[2];
            v[3] 0 -v[1];
            -v[2] v[1] 0]
end

function unhat(S)
    return 0.5*[S[3,2]-S[2,3];
                S[1,3]-S[3,1];
                S[2,1]-S[1,2]]
end

H = [zeros(1,3); I];
T = [1  zeros(1,3);
     zeros(3,1) -I];

function L(q)
    return [q[1]          -q[2:4]';
            q[2:4]    q[1]*I + hat(q[2:4])]
end

function R(q)
    return [q[1]          -q[2:4]';
            q[2:4]    q[1]*I - hat(q[2:4])]
end

function G(q)
    return L(q)*H
end

function Q(q)
    return H'*(R(q)'*L(q))*H
end

function expq(ϕ)
    θ = norm(ϕ)
    return [cos(θ); ϕ*sinc(θ/π)];
end

function logq(q)
    θ = acos(q[1])
    r = q[2:4]/norm(q[2:4])
    return θ*r
end
```

```julia
#dynamics
function dynamics(x)
    q = x[1:4]
    q = q/norm(q)
    
    ω = x[5:7]
    
    q̇ = 0.5*G(q)*ω
    ω̇ = -J\(hat(ω)*J*ω)

    ẋ = [q̇; ω̇]
end
```

```julia
#Classic RK4 integrator: https://en.wikipedia.org/wiki/Runge%E2%80%93Kutta_methods
function rkstep(x)
    f1 = dynamics(x)
    f2 = dynamics(x + 0.5*h*f1)
    f3 = dynamics(x + 0.5*h*f2)
    f4 = dynamics(x + h*f3)
    xn = x + (h/6.0)*(f1 + 2*f2 + 2*f3 + f4)
    xn[1:4] .= xn[1:4]/norm(xn[1:4]) #re-normalize quaternion
    return xn
end
```

```julia
J = Diagonal([1.0; 1.25; 1.5])
```

```julia
h = 0.1 #time step
n = 600 #number of time steps
tf = n*h #final time

#sample random initial attitude
q0 = randn(4)
q0 = q0/norm(q0)

#sample random angular velocity
ω0 = 0.1*randn(3)

#sample random gyro bias
β0 = 0.3*randn(3)

x0 = [q0; ω0];
```

```julia
#Simulate n time steps
xtraj = zeros(7,n)
xtraj[:,1] .= x0
for k = 1:(n-1)
    xtraj[:,k+1] = rkstep(xtraj[:,k])
end
```

```julia
# Noise
m = 2 #number of star trackers
W = 0.001*I(3*m) #measurement noise
V = 0.01*I(6) #process noise
```

```julia
#generate noisy gyro measurements
gyro = zeros(3,n)
for k = 1:n
    gyro[:,k] = xtraj[5:7] + β0 + sqrt(V[4:6,4:6])*randn(3)
end
```

```julia
#Generate noisy star-tracker measurements
ytraj = zeros(4*m,n)
for k = 1:n
    qk = xtraj[1:4,k]
    yk = zeros(4,m)
    wk = reshape(sqrt(W)*randn(3*m),3,m)
    for ℓ = 1:m
        yk[:,ℓ] = L(qk)*expq(wk[:,ℓ])
    end
    ytraj[:,k] .= yk[:]
end
```

```julia
function state_prediction(x,u,h)
    q = x[1:4]
    β = x[5:7]
    ω = u-β
    Δq = expq(0.5*h*ω)
    return [L(q)*Δq; β]
end

function state_prediction_deriv(x,u,h)
    q = x[1:4]
    β = x[5:7]
    ω = u-β
    Δq = expq(0.5*h*ω)
    qn = L(q)*Δq
    A = [G(qn)'*R(Δq)*G(q) -0.5*h*G(qn)'*G(q);
         zeros(3,3)            I(3)]
    return A
end
```

```julia
function innovation(x,y)
    q = x[1:4]
    zk = zeros(eltype(x),3,m)
    yk = reshape(y,4,m)
    for k = 1:m
        zk[:,k] = G(q)'*yk[:,k]
    end
    return zk[:]
end

function innovation_deriv(x,y)
    q = x[1:4]
    C = zeros(3*m,6)
    yk = reshape(y,4,m)
    for k = 1:m
        C[((k-1)*3).+(1:3),1:3] .= H'*R(yk[:,k])*T*G(q)
    end
    return C
end
```

```julia
# Filter Initialization
xfilt = zeros(7, n)
xfilt[1:4,1] = q0+0.01*randn(4)
xfilt[1:4,1] = xfilt[1:4,1]/norm(xfilt[1:4,1])

P = zeros(6,6,n)
P[:,:,1] .= 0.5*I(6);
```

```julia
for k = 1:(n-1)
    #Prediction
    xpred = state_prediction(xfilt[:,k],gyro[:,k],h)
    A = state_prediction_deriv(xfilt[:,k],gyro[:,k],h)
    Ppred = A*P[:,:,k]*A' + V

    #Innovation
    z = innovation(xpred, ytraj[:,k+1])
    C = innovation_deriv(xpred, ytraj[:,k+1])
    S = C*Ppred*C' + W

    #Kalman Gain
    K = Ppred*C'/S

    #Update
    Δx = -K*z
    ϕ = Δx[1:3]
    Δq = [sqrt(1-ϕ'*ϕ); ϕ]
    xfilt[:,k+1] = [L(xpred[1:4])*Δq; xpred[5:7]+Δx[4:6]]
    
    P[:,:,k+1] = (I(6) - K*C)*Ppred*(I(6) - K*C)' + K*W*K'
end
```

```julia
##Consistency Plots - we want the errors to be within the 2-sigma bounds 95% of the time
```

```julia
ϕ_error = zeros(3,n)
b_error = zeros(3,n)
σ_bounds = zeros(6,n)
for k = 1:n
    ϕ_error[:,k] = logq(L(xtraj[1:4,k])'*xfilt[1:4,k])
    b_error[:,k] = xfilt[5:7,k] - xtraj[5:7,k]
    σ_bounds[:,k] = sqrt.(diag(P[:,:,k]))
end
```
