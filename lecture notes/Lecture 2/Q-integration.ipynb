```julia
using LinearAlgebra
```

```julia
function hat(v)
    return [0 -v[3] v[2];
            v[3] 0 -v[1];
            -v[2] v[1] 0]
end
```

```julia
h = 0.1
```

```julia
#initial conditions
Q0 = Array(I(3))
ω0 = 0.1*randn(3)
x0 = [vec(Q0); ω0];
```

```julia
#dynamics
function dynamics(x)
    Q = reshape(x[1:9],3,3)
    ω = x[10:12]
    
    Q̇ = Q*hat(ω)
    ω̇ = randn(3)

    ẋ = [vec(Q̇); ω̇]
end
```

```julia
function rkstep(x)
    f1 = dynamics(x)
    f2 = dynamics(x + 0.5*h*f1)
    f3 = dynamics(x + 0.5*h*f2)
    f4 = dynamics(x + h*f3)
    xn = x + (h/6.0)*(f1 + 2*f2 + 2*f3 + f4)
    return xn
end
```

```julia
xk = x0
for k = 1:10000
    xk = rkstep(xk)
end
```

```julia
Qk = reshape(xk[1:9],3,3)
```
