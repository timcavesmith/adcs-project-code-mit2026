```julia
using LinearAlgebra
```

```julia
function hat(v)
    return [0 -v[3] v[2];
            v[3] 0 -v[1];
            -v[2] v[1] 0]
end
```

```julia
#initial conditions
#Q0 = Array(I(3))
Q0 = exp(hat(randn(3))) #sample a random rotation matrix for initial attitude
ω0 = 0.1*randn(3)
```

```julia
#Constant velocity
h = 0.1 #time step
n = 1000 #number of time steps
T = n*h #final time
```

```julia
#Assume constant velocity (exact solution)
Q = Q0*exp(hat(T*ω0))
#Q = Q0*(I + hat(T*ω0)) #Linear approximation
```

```julia
Q'*Q
```

```julia
det(Q)
```

```julia
#dynamics
function dynamics(x)
    Q = reshape(x[1:9],3,3)
    ω = x[10:12]
    
    Q̇ = Q*hat(ω)
    #ω̇ = zeros(3)
    ω̇ = randn(3) #random walk

    ẋ = [vec(Q̇); ω̇]
end
```

```julia
#Classic RK4 integrator: https://en.wikipedia.org/wiki/Runge%E2%80%93Kutta_methods
function rkstep(x)
    f1 = dynamics(x)
    f2 = dynamics(x + 0.5*h*f1)
    f3 = dynamics(x + 0.5*h*f2)
    f4 = dynamics(x + h*f3)
    xn = x + (h/6.0)*(f1 + 2*f2 + 2*f3 + f4)
    return xn
end
```

```julia
#Simulate n time steps
x0 = [vec(Q0); ω0];
xk = x0
for k = 1:n
    xk = rkstep(xk)
end
```

```julia
#Extract final rotation matrix
Qn = reshape(xk[1:9],3,3)
```
