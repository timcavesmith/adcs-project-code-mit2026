```julia
using LinearAlgebra
```

```julia
function hat(v)
    return [0 -v[3] v[2];
            v[3] 0 -v[1];
            -v[2] v[1] 0]
end
function unhat(S)
    return 0.5*[S[3,2]-S[2,3];
                S[1,3]-S[3,1];
                S[2,3]-S[3,2]]
end
```

```julia
#Make a random attitude
Qtrue = exp(hat(randn(3)))
```

```julia
#Make some random vectors
n = 3
r_B = randn(3,n)
for k = 1:n
    r_B[:,k] .= r_B[:,k]/norm(r_B[:,k])
end

r_N = Qtrue*r_B
```

```julia
#Add some noise
for k = 1:n
    r_B[:,k] += 0.1*randn(3)
    r_B[:,k] .= r_B[:,k]/norm(r_B[:,k])
end
```

```julia
#Attitude Profile Matrix
w = ones(n)
B = zeros(3,3)
for k = 1:n
    B += w[k]*r_B[:,k]*r_N[:,k]'
end
```

```julia
using Convex, Clarabel
```

```julia
#Set up optimization problem
Q = Variable(3,3)
prob = maximize(tr(B*Q), [I(3) Q'; Q I(3)] ⪰ 0)
```

```julia
solve!(prob, Clarabel.Optimizer)
```

```julia
Q1 = Q.value
```

```julia
#Is the solution orthogonal?
Q1'*Q1
```

```julia
#Is it "special"?
det(Q1)
```

```julia
#Does it match the true solution?
Q1'*Qtrue
```

```julia
#Error in degrees
(180/pi)*norm(unhat(log(Q1'*Qtrue)))
```
