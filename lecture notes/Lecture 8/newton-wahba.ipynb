```julia
using LinearAlgebra
using ForwardDiff
```

```julia
#quaternion functions
function hat(v)
    return [0 -v[3] v[2];
            v[3] 0 -v[1];
            -v[2] v[1] 0]
end

function unhat(S)
    return 0.5*[S[3,2]-S[2,3];
                S[1,3]-S[3,1];
                S[2,1]-S[1,2]]
end

H = [zeros(1,3); I];
T = [1  zeros(1,3);
     zeros(3,1) -I];

function L(q)
    return [q[1]          -q[2:4]';
            q[2:4]    q[1]*I + hat(q[2:4])]
end

function R(q)
    return [q[1]          -q[2:4]';
            q[2:4]    q[1]*I - hat(q[2:4])]
end

function G(q)
    return L(q)*H
end

function Q(q)
    return H'*(R(q)'*L(q))*H
end

function expq(ϕ)
    θ = norm(ϕ)
    return [cos(θ); ϕ*sinc(θ/π)];
end
```

```julia
#Make a random attitude
qtrue = expq(0.5*randn(3))
Qtrue = Q(qtrue) #equivalent rotation matrix
```

```julia
#Make some random vectors
n = 3
r_B = randn(3,n)
for k = 1:n
    r_B[:,k] .= r_B[:,k]/norm(r_B[:,k])
end

r_N = Qtrue*r_B
```

```julia
#Add some noise
for k = 1:n
    r_B[:,k] += 0.1*randn(3)
    r_B[:,k] .= r_B[:,k]/norm(r_B[:,k])
end
```

```julia
function residual(q)
    s = r_N - Q(q)*r_B
    return s[:]
end
```

```julia
#Random initial guess
q = randn(4)
q = q/norm(q)
```

```julia
#Gauss-Newton Method
ϕ = ones(3)
iter = 0
while maximum(abs.(ϕ)) > 1e-8
    s = residual(q)
    ds = ForwardDiff.jacobian(residual, q)
    ∇s = ds*G(q)
    ϕ = -(∇s'*∇s)\(∇s'*s) #3-parameter update computed with gauss-newton
    q = L(q)*[sqrt(1-ϕ'*ϕ); ϕ] #multiplicative update applied to q
    iter += 1
end
```
