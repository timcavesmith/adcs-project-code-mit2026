```julia
using LinearAlgebra
```

```julia
#quaternion functions
function hat(v)
    return [0 -v[3] v[2];
            v[3] 0 -v[1];
            -v[2] v[1] 0]
end

function unhat(S)
    return 0.5*[S[3,2]-S[2,3];
                S[1,3]-S[3,1];
                S[2,1]-S[1,2]]
end

H = [zeros(1,3); I];
T = [1  zeros(1,3);
     zeros(3,1) -I];

function L(q)
    return [q[1]          -q[2:4]';
            q[2:4]    q[1]*I + hat(q[2:4])]
end

function R(q)
    return [q[1]          -q[2:4]';
            q[2:4]    q[1]*I - hat(q[2:4])]
end

function G(q)
    return L(q)*H
end

function Q(q)
    return H'*(R(q)'*L(q))*H
end

function expq(ϕ)
    θ = norm(ϕ)
    return [cos(θ); ϕ*sinc(θ/π)];
end
```

```julia
#Make a random attitude
qtrue = expq(0.5*randn(3))
Qtrue = Q(qtrue) #equivalent rotation matrix
```

```julia
#Make some random vectors
n = 3
r_B = randn(3,n)
for k = 1:n
    r_B[:,k] .= r_B[:,k]/norm(r_B[:,k])
end

r_N = Qtrue*r_B
```

```julia
#Add some noise
for k = 1:n
    r_B[:,k] += 0.1*randn(3)
    r_B[:,k] .= r_B[:,k]/norm(r_B[:,k])
end
```

```julia
#Davenport matrix
w = (1/n)*ones(n)
D = zeros(4,4)
for k = 1:n
    D += w[k]*L(H*r_N[:,k])'*R(H*r_B[:,k])
end
```

```julia
eigvals(D)
```

```julia
eigvecs(D)
```

```julia
q = eigvecs(D)[:,4]
```
