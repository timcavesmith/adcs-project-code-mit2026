```julia
using LinearAlgebra
```

```julia
#Projectile model

g = 9.8

#initial conditions
z0 = randn() 
v0 = randn()

#measurement times
t1 = 1.0
t2 = 2.0

#measurements
z1 = z0 + v0*t1 - 0.5*g*t1*t1
z2 = z0 + v0*t2 - 0.5*g*t2*t2;
```

```julia
#estimation problem
C = [1 t1;
     1 t2]

y = [z1; z2] + [0.5*g*t1*t1; 0.5*g*t2*t2]

x0 = C\y
```

```julia
#Try with more noisy measurements
m = 20
t = randn(m)

W = 0.1*I(m)

#w = zeros(m)
w = sqrt(W)*randn(m)

y = z0*ones(m) .+ v0.*t .+ w

C = [ones(m) t]
```

```julia
x̄ = (C'*inv(W)*C)\(C'*inv(W)*y)
```

```julia
#Recursive updates

#start with just the first two measurements
C = [ones(2) t[1:2]]
P = inv(C'*inv(W[1:2,1:2])*C)
x̄_recursive = P*(C'*inv(W[1:2,1:2])*y[1:2])
```

```julia
for k = 3:m
    Ck = [1 t[k]]
    Pk = inv(inv(P) + Ck'*inv(W[k,k])*Ck)
    x̄_recursive .= Pk*(Ck'*inv(W[k,k])*y[k] + inv(P)*x̄_recursive)
    P = Pk
end
```
