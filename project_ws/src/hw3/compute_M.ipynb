```python
# Find M for gyro
from scipy.spatial.transform import Rotation
import numpy as np
import sensors

# Misalignment Error
q0 = Rotation.from_euler('xyz', [0.00416667, -0.00416667, 0.00416667], degrees=True).as_quat()
R0 = Rotation.from_quat(q0).as_matrix()

# Scale Factor Error
scale_factor_error = 15.45/1000000
R0[0] = R0[0] - scale_factor_error*R0[0]
R0[1] = R0[1] - scale_factor_error*R0[1]
R0[2] = R0[2] + scale_factor_error*R0[2]
R0
```
